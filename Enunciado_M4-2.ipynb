* Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente
* Mostrar cuantos papeles para actores hubo en la década de los 50's
* Mostrar cuantos papeles para actrices hubo en la década de los 50's

```python
import pandas as pd 
def load_and_process_data(file_path_escritorio_actividad_9):
    df = pd.read_csv(file_path_escritorio_actividad_9)

titulos = pd.read_csv("imdb_titulos.csv")
elenco = pd.read_csv("imdb_elenco.csv")

# 1) Mostrar 5 primeros registros de cada uno
print("=== titulos: primeros 5 ===")
print(titulos.head())

print("\n=== elenco: primeros 5 ===")
print(elenco.head())

# 2) Número de registros
print("\nNúmero de registros en titulos:", len(titulos))
print("Número de registros en elenco:", len(elenco))

# 3) 5 películas más antiguas
print("\n=== 5 películas más antiguas ===")
print(titulos.sort_values("year").head(5))

# 4) Películas con 'Dracula' en el título
dracula = titulos[titulos["title"].str.contains("Dracula", case=False, na=False)]
print("\n=== Películas con 'Dracula' en el título ===")
print(dracula)
print("Total de películas con 'Dracula':", len(dracula))

# 5) Los 10 títulos más comunes
print("\n=== 10 títulos más comunes ===")
print(titulos["title"].value_counts().head(10))

# 6) Primera película titulada 'Romeo and Juliet'
romeo = titulos[titulos["title"] == "Romeo and Juliet"].sort_values("year")
print("\n=== Primera 'Romeo and Juliet' ===")
print(romeo.head(1))

# 7) Películas con 'Exorcist' ordenadas de la más vieja a la más reciente
exorcist = titulos[titulos["title"].str.contains("Exorcist", case=False, na=False)] \
    .sort_values("year")
print("\n=== Películas con 'Exorcist' ===")
print(exorcist)

# 8) Cuántas películas se hicieron en 1950
pelis_1950 = titulos[titulos["year"] == 1950]
print("\nPelículas hechas en 1950:", len(pelis_1950))

# 9) Cuántas películas se hicieron de 1950 a 1959
pelis_50s = titulos[(titulos["year"] >= 1950) & (titulos["year"] <= 1959)]
print("Películas hechas entre 1950 y 1959:", len(pelis_50s))

# 10) Todos los roles en "The Godfather"
godfather_roles = elenco[elenco["title"] == "The Godfather"]
print("\n=== Roles en 'The Godfather' ===")
print(godfather_roles)
print("Total de roles en 'The Godfather':", len(godfather_roles))

# 11) Elenco de "Dracula" (1958) ordenado por 'n'
dracula_1958 = elenco[(elenco["title"] == "Dracula") & (elenco["year"] == 1958)] \
    .sort_values("n")
print("\n=== Elenco de 'Dracula' (1958) ordenado por 'n' ===")
print(dracula_1958)

# 12) Cuántos papeles de "Bruce Wayne" se han hecho
bruce_wayne = elenco[elenco["character"].str.contains("Bruce Wayne", case=False, na=False)]
print("\nPapeles de 'Bruce Wayne' en la historia del cine:", len(bruce_wayne))

# 13) Cuántos papeles ha hecho "Robert De Niro"
deniro = elenco[elenco["name"] == "Robert De Niro"]
print("Papeles de 'Robert De Niro':", len(deniro))

# 14) Papeles protagonistas (n=1) de Charlton Heston en los 60, ordenados por año desc
heston_60s = elenco[
    (elenco["name"] == "Charlton Heston") &
    (elenco["n"] == 1) &
    (elenco["year"] >= 1960) &
    (elenco["year"] <= 1969)
].sort_values("year", ascending=False)

print("\n=== Protagonismos de Charlton Heston en los 60 (n=1) ===")
print(heston_60s[["title", "year", "character", "n"]])

# 15) Cuántos papeles para actores hubo en los 50
# (asumiendo columna 'type' con valores 'actor' / 'actress')
actores_50s = elenco[
    (elenco["type"] == "actor") &
    (elenco["year"] >= 1950) &
    (elenco["year"] <= 1959)
]
print("\nPapeles para actores en los 50:", len(actores_50s))

# 16) Cuántos papeles para actrices hubo en los 50
actrices_50s = elenco[
    (elenco["type"] == "actress") &
    (elenco["year"] >= 1950) &
    (elenco["year"] <= 1959)
]
print("Papeles para actrices en los 50:", len(actrices_50s))

```
